```python
#we do the same but woth the segmentation
#read sitk image
time = 1
rad = 'L'
type = 'G'
im_path = repo_path / 'data/segmentations'/f'{rad}_{time}_seg/2_{type}_L.seg.nrrd'
im_sitk = sitk.ReadImage(str(im_path))
im_np = sitk.GetArrayFromImage(im_sitk)
#get image origin
origin = im_sitk.GetOrigin()
print(f'origin is: {origin}')
#get direction
direction = im_sitk.GetDirection()
print(f'direction is: {direction}')
plt.figure(figsize=(6,6))
plt.imshow(im_np[0], cmap='gray')
#flip
im_sitk_flip = sitk.Flip(im_sitk, [True, False, False])
#plot change
plt.figure(figsize=(6,6))
im_np_flip = sitk.GetArrayFromImage(im_sitk_flip)
plt.imshow(im_np_flip[0], cmap='gray')
#origin and direction are not changed
origin = im_sitk_flip.GetOrigin()
print(f'origin is: {origin}')
direction = im_sitk_flip.GetDirection()
print(f'direction is: {direction}')
#save flipped image
fliped_path = repo_path / 'data/segmentations'/f'{rad}_{time}_seg/2_{type}_L_flip.seg.nrrd'
sitk.WriteImage(im_sitk_flip, str(fliped_path))
```
